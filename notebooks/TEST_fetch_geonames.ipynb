```python
import requests as rq
import pandas as pd
```

```python
GEONAMES_API = "http://api.geonames.org/childrenJSON"
payload = {"geonameId": "1269750", "username": "prathafish"}

resp = rq.get(GEONAMES_API, params=payload)
resp.json()
```

```python
resp_json = resp.json()['geonames']
columns = ["geonameId", "name", "adminCodes1", "countryName", "countryId"]
resp_trim = [{k: v for k, v in d.items() if k in columns} for d in resp_json]

pd.DataFrame(resp_trim)
```
