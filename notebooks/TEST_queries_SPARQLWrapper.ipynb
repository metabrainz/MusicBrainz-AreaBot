### This notebook tests some relevant wikidata queries for the AreaBot

```python
from SPARQLWrapper import SPARQLWrapper, JSON
import pandas as pd
sparql = SPARQLWrapper("https://query.wikidata.org/sparql")
```

```python
def fetch_query(query):
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()
    
    return results

def results_to_df(results):

    results_df = pd.json_normalize(results['results']['bindings'])
    results_df = results_df[[col for col in results_df.columns if col.endswith('value')]]
    results_df.columns = [col.split('.')[0] for col in results_df.columns]
    return results_df
```

```python
%%time
query1 = """SELECT ?country ?countryLabel ?iso_3166_1_code WHERE {
  ?country wdt:P31 wd:Q3624078.
  ?country wdt:P297 ?iso_3166_1_code.
  SERVICE wikibase:label { bd:serviceParam wikibase:language "en". }
}
"""

result1 = fetch_query(query1)

print(result1, "\n")

df1 = results_to_df(result1)
df1
```
